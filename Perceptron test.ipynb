### Algorithms

```python
def winnow(points, labels, m, eta):
    d = points.shape[1]
    all_correct = False
    w = np.ones(m)/d
    count = 0
    while not all_correct:
        all_correct = True
        for i in range(m):
            x = points[i,:]
            y = labels[i]
            y_hat = np.sign(np.dot(w,x)+1e-32)
            if y!=y_hat:
                for j in range(d):
                    w[i] *= np.exp(eta*y*x[i])
                    count += 1
                w = w/(np.sum(w))
                all_correct = False
    
    return w, count
```

```python
def perceptron(points, labels):
    n, d = points.shape
    
    all_correct = False
    w = np.zeros(d)
    
    total_count = 0         # Total iteration count
    pass_count = 0          # How many passes through the entire data
    mistake_count = 0       # Number of mistakes 
    
    while not all_correct:
        all_correct = True
        for i in range(n):
            x = points[i]
            y = labels[i]
            y_hat = np.sign(w@x+1e-32)
            if y!=y_hat:
                w += y*x
                all_correct = False
                mistake_count += 1            
            total_count += 1            
        pass_count += 1
        
    # Adjust for the last pass
    total_count -= n
    pass_count -= 1
    
    # Average number of mistakes per pass through the data
    if mistake_count == 0:
        avg_mistakes = 0
    else:
        avg_mistakes = mistake_count/pass_count
            
    return w, total_count, avg_mistakes
```

```python
# Choosing a random point for updates
def random_perceptron(points, labels):
    n, d = points.shape
    
    all_correct = False
    w = np.zeros(d)
    
    total_count = 0         # Total iteration count
    mistake_count = 0       # Number of mistakes 
    
    while not all_correct:
        all_correct = True
        i = np.random.randint(n)
        x = points[i]
        y = labels[i]
        y_hat = np.sign(w@x+1e-32)
        if y!=y_hat:
            w += y*x
            mistake_count += 1
        total_count += 1
        
        # Check if separator was found
        if np.any(np.sign(points@w+1e-32)!=labels):
            all_correct = False
    
    # Average number of mistakes every n points
    avg_count = n*mistake_count/total_count
    
    return w, total_count, avg_count
```

```python
# def comp_updt(points, labels, w, r, p_curr):
#     num = p_curr*np.exp(-labels*(points@w)/(r**2))
#     return num/np.sum(num)

def opt_sep(points, labels):
    n, d = points.shape
    
    r = np.max(np.linalg.norm(points, axis=1))
    
    all_correct = False
    p_prev = np.ones(n)/n   # p_(t-2)
    p_curr = np.ones(n)/n   # p_(t-1)
    w = np.zeros(d)
    w_sum = np.zeros(d)
        
    total_count = 0         # Total iteration count
    pass_count = 0          # How many passes through the entire data
    
    while not all_correct:
        all_correct = True
        
        # Learner update
        w += ((2*p_curr-p_prev)*labels)@points
                
        # Data update
        p_prev = p_curr.copy()
        
        num = p_curr*np.exp(-labels*(points@w)/(r**2))
        p_curr = num/np.sum(num)
        
#         p_curr = comp_updt(points, labels, w, r, p_curr)
    
        total_count += 1
        
#         for i in range(n):
#             x = points[i]
#             y = labels[i]
#             p_curr[i] *= np.exp(-(1/(r**2))*y*w@x)
#             total_count += 1
#         p_curr = p_curr/np.sum(p_curr)
        
        # Check if average w separates data
        pass_count += 1
        w_sum += w
        avg_w = w_sum/pass_count
        
        if np.any(np.sign(points@avg_w+1e-32)!=labels):        
            all_correct = False


#     print('n:', n)
#     print('logn:', np.log(n))
#     print('pT', np.min(np.log(1/p_curr)))
        
    return avg_w, total_count
```

```python
def x_mu(points, mu, y):
    num = np.exp(-points.T@y/mu)
    return num/np.sum(num)

def smoothed_perceptron(points):
    D, N = points.shape
    
    y = points@np.ones(N)/N
    mu = 1
    x = x_mu(points, mu, y)
    
    x_prev = x.copy()
    
    all_correct = False    
    k = 0
    
    while not all_correct:
        all_correct = True
        
        theta = 2/(k+3)

        # Original SP
#         y = (1-theta)*(y+theta*points@x) + (theta**2)*points@x_mu(points, mu, y)
#         mu = (1-theta)*mu
#         x = (1-theta)*x + theta*x_mu(points, mu, y)

        # Simplified SP
#         y = y + points@x + points@x_mu(points, 1, y)
        y += 2*points@x - points@x_prev
        x_prev = x.copy()
        x += x_mu(points, 1, y)
        
        k += 1
        
        if np.any(points.T@y <= 0):        
            all_correct = False
            
    return y, k
```

We compare the total number of iterations needed by each algorithm to find a perfect separator, as well as the time until convergence. For Perceptron, this also includes iterations in which a mistake wasn't made. (Even if we only account for mistakes on Perceptron, the difference between both algorithms is significant).
